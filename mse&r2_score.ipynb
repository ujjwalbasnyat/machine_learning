```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/dataset_container/placement_second.csv')
df.head()
```

```python
sns.scatterplot(x=df['cgpa'], y=df['package'])
```

```python
X= df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
sns.scatterplot(x=df['cgpa'], y=df['package'])
plt.plot(X_train, lr.predict(X_train), color='red')
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
print('mean sqaure error:', mean_squared_error(y_test, lr.predict(X_test)))
```

```python
print('rmse:', np.sqrt(mean_squared_error(y_test, lr.predict(X_test))))
```

```python
print('r2 score:', r2_score(y_test, lr.predict(X_test)))
```
