```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/dataset_container/titanic.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.drop(columns = ['Name','Ticket','Cabin', 'PassengerId'], inplace=True)
```

```python
df.head()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Fare'].fillna(df['Fare'].mean(), inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Survived', axis=1), df['Survived'], test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.compose import ColumnTransformer
```

```python
Transformer = ColumnTransformer(transformers = [
    ('minmax', MinMaxScaler(),['Age', 'Fare']),
    ('ohe', OneHotEncoder(sparse_output=False, drop ='first'), ['Sex', 'Embarked'])
], remainder = 'passthrough')
```

```python
X_train_transformed = Transformer.fit_transform(X_train)
X_test_transformed = Transformer.transform(X_test)
```

```python
train_df = pd.DataFrame(X_train_transformed)
test_df = pd.DataFrame(X_test_transformed)
```
